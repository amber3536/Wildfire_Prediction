### Alberta weather data located here?

```python
import requests
import lxml
from bs4 import BeautifulSoup
```

```python
url = "https://dd.weather.gc.ca/climate/observations/daily/csv/AB/"
f = requests.get(url)
```

```python
soup = BeautifulSoup(f.text, 'html.parser')
```

```python
s = "climate_daily_AB_3031676_1968-12_P1D.csv"
s = s.split('_')[-2]
s = s.split('-')[0]
s
```

```python
links = soup.find_all("a")[5:] # Find all elements with the tag <a>
l_links = []

for link in links:
  s = link.string
  s = s.split('_')[-2]
  s = s.split('-')[0]
  if s < '2019' and s > '1989':
    l_links.append(link.get("href"))
  #print("Link:", link.get("href"))
```

```python
u = url + l_links[0]
```

```python
f = requests.get(u)
```

```python
soup = BeautifulSoup(f.text, 'html.parser')
```

```python
col = ["Longitude (x)","Latitude (y)","Station Name","Climate ID","Date/Time","Year","Month","Day","Data Quality","Max Temp (C)","Max Temp Flag","Min Temp (C)","Min Temp Flag","Mean Temp (C)","Mean Temp Flag","Heat Deg Days (C)","Heat Deg Days Flag","Cool Deg Days (C)","Cool Deg Days Flag","Total Rain (mm)","Total Rain Flag","Total Snow (cm)","Total Snow Flag","Total Precip (mm)","Total Precip Flag","Snow on Grnd (cm)","Snow on Grnd Flag","Dir of Max Gust (10s deg)","Dir of Max Gust Flag","Spd of Max Gust (km/h)","Spd of Max Gust Flag"]
df = pd.read_csv(u, names=col, encoding = "ISO-8859-1")
```
